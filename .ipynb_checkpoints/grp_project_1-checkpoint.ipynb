```python
import pandas as pd
import numpy as np
from pathlib import Path

# ========== 配置 ==========
BASE_PATH = Path("/Users/wantwa/.cache/kagglehub/datasets/mohammedobeidat/us-stock-market-data/versions/2")
TICKERS = ['AAPL', 'MSFT', 'GOOGL', 'AMZN', 'JNJ', 'XOM', 'V', 'WMT', 'NVDA', 'PG',
           'LLY', 'CVX', 'MA', 'HD', 'PFE', 'MRK', 'KO', 'PEP', 'ORCL',
           'TMO', 'AZN', 'CSCO', 'DHR', 'MCD', 'ABT', 'TMUS', 'ACN', 'NEE', 'VZ', 'TTE',
           'LIN', 'DIS', 'PM', 'BMY', 'CMCSA', 'SCHW', 'UPS', 'TXN', 'RTX', 'COP']

# Forecasting the return by future 21 trading days (1 month)
FORECAST_HORIZON = 21

# Fundamental lag month -- 3 month for one quarter
FUND_LAG_MONTHS = 3
```

# 2. Feature Engineering

```python
def load_single_stock(ticker, base_path, data_type="train"):
    """
    loading fund. & tech. data of single stock, adding ticker label column
    """
    # Technical
    tech_path = base_path / "technials" / data_type / f"{ticker}_technicals_{data_type}.csv"
    tech_df = pd.read_csv(tech_path, parse_dates=['Date'])
    tech_df['ticker'] = ticker
    tech_df.set_index('Date', inplace=True)

    # Fundamental
    fund_path = base_path / "fundamentals" / data_type / f"{ticker}_fundamentals_{data_type}.csv"
    fund_df = pd.read_csv(fund_path, parse_dates=['Date'])
    fund_df['ticker'] = ticker
    fund_df.set_index('Date', inplace=True)

    # Left Join
    merged = tech_df.join(fund_df, how='left', rsuffix='_fund')
    # Delete duplicate ticker columns
    merged.drop(columns=['ticker_fund'], inplace=True, errors='ignore')

    # Reset Index -- Let 'Date' as normal column for convenient sorting
    merged.reset_index(inplace=True)
    return merged

def build_panel_dataset(tickers, base_path, data_type="train"):
    all_dfs = []
    for ticker in tickers:
        try:
            df = load_single_stock(ticker, base_path, data_type)
            all_dfs.append(df)
            print(f"✓ {ticker}")
        except Exception as e:
            print(f"✗ {ticker}: {e}")

    panel = pd.concat(all_dfs, ignore_index=True)
    # Sort by Date and ticker
    panel = panel.sort_values(['Date', 'ticker']).reset_index(drop=True)
    return panel

def lag_fundamentals(panel, lag_months=3):
    """
    Shift the fundamentals column forward by 'lag_months' months (by date).
    Note: Shifting is independent for each stock.
    """
    fund_cols = ['pe_ratio', 'pb_ratio', 'ps_ratio', 'pfcf_ratio',
                 'return_on_equity', 'return_on_assets', 'debt_to_equity_ratio',
                 'current_ratio', 'ebitda', 'revenue']
    # The actual column names may not have a suffix; we will use these names directly.
    available_fund_cols = [c for c in fund_cols if c in panel.columns]

    # For each stock, shift the date forward by lag_months months.
    def shift_group(df_group):
        df_group = df_group.sort_values('Date')
        for col in available_fund_cols:
            df_group[col] = df_group[col].shift(lag_months * 30)
        return df_group

    panel = panel.groupby('ticker', group_keys=False).apply(shift_group)
    return panel

def compute_features(panel):
    """
    Explicitly compute each feature based on your selected feature list.
    """
    # Calculated by stock group
    panel = panel.sort_values(['ticker', 'Date']).copy()

    # 1. Return: Past 1 month (21 days) and 3 months (63 days)
    panel['return_1m'] = panel.groupby('ticker')['close'].pct_change(periods=21)
    panel['return_3m'] = panel.groupby('ticker')['close'].pct_change(periods=63)

    # 2. Momentum over 12 months (approximately 252 trading days)
    panel['momentum_12m'] = panel.groupby('ticker')['close'].pct_change(periods=252)

    # 3. sma_50/sma_200 ratio (already in dataset 'sma_50-sma_200_rate') -- use directly
    panel['sma_50_sma_200_ratio'] = panel['sma_50-sma_200_rate']

    # 4. MACD histogram ('MACDh_10_26_9')
    panel['MACDh'] = panel['MACDh_10_26_9']

    # 5. Volatility in 1 month (using 'stdev')
    panel['volatility_1m'] = panel['stdev']

    # 6. ATR
    panel['ATR'] = panel['atr']

    # 7. Log trading volume
    panel['log_volume'] = np.log1p(panel['volume'])

    # 8. OBV
    panel['obv'] = panel['obv']

    # 9. Fundamental feature (already processed by function: lag_fundamentals)
    panel['pe_ratio'] = panel['pe_ratio']
    panel['pb_ratio'] = panel['pb_ratio']
    panel['ps_ratio'] = panel['ps_ratio']
    panel['pfcf_ratio'] = panel['pfcf_ratio']
    panel['return_on_equity'] = panel['return_on_equity']
    panel['return_on_assets'] = panel['return_on_assets']
    panel['debt_to_equity'] = panel['debt_to_equity_ratio']
    panel['current_ratio'] = panel['current_ratio']

    # 10. EBITDA margin = ebitda / revenue
    panel['ebitda_margin'] = panel['ebitda'] / panel['revenue']

    return panel

def generate_regression_target(panel, horizon=21):
    """
    Generate regression target: Future horizon day(21) continuous return (decimal)
    """
    panel = panel.sort_values(['ticker', 'Date']).copy()
    panel['future_return'] = panel.groupby('ticker')['close'].pct_change(periods=horizon).shift(-horizon)

    return panel

def handle_missing(panel, feature_cols):
    """
    Forward fill feature columns within each stock, but never fill the target.
    This avoids leaking the last available future return into rows that should be dropped.
    """
    panel = panel.sort_values(['ticker', 'Date']).copy()
    panel[feature_cols] = panel.groupby('ticker')[feature_cols].ffill()
    panel = panel.dropna(subset=feature_cols + ['future_return']).reset_index(drop=True)
    return panel
```

## main function

```python
# 1. Build Panel dataset
print("Building train data...")
train_raw = build_panel_dataset(TICKERS, BASE_PATH, "train")
print(f"Raw trainig set shape: {train_raw.shape}")

print("Building test data...")
test_raw = build_panel_dataset(TICKERS, BASE_PATH, "test")
print(f"Raw testing set shape: {test_raw.shape}")

# 2. Fund. lag (avoid data leakage)
print("Starting processing fundamental lag...")
train_raw = lag_fundamentals(train_raw, lag_months=FUND_LAG_MONTHS)
test_raw = lag_fundamentals(test_raw, lag_months=FUND_LAG_MONTHS)

# 3. Feature Engineering
print("Computing Features...")
train_df = compute_features(train_raw)
test_df = compute_features(test_raw)

# 4. Generate forcasting y
print("Generate forcasting target...")
train_df = generate_regression_target(train_df, horizon=FORECAST_HORIZON)
test_df = generate_regression_target(test_df, horizon=FORECAST_HORIZON)

# 5. Our feature selected
feature_cols = [
    'return_1m', 'return_3m', 'momentum_12m', 'sma_50_sma_200_ratio',
    'MACDh', 'volatility_1m', 'ATR', 'log_volume', 'obv',
    'pe_ratio', 'pb_ratio', 'ps_ratio', 'pfcf_ratio',
    'return_on_equity', 'return_on_assets', 'ebitda_margin',
    'debt_to_equity', 'current_ratio'
]

# Ensure all feature columns exist
missing_features = [f for f in feature_cols if f not in train_df.columns]
if missing_features:
    print(f"Warning: These feature are missing {missing_features}")
    # hands-on create the missing column
    for f in missing_features:
        train_df[f] = np.nan
        test_df[f] = np.nan

# 6. Processing Missing Value
print("Processing Missing Value...")
train_df = handle_missing(train_df, feature_cols)
test_df = handle_missing(test_df, feature_cols)

# 7. Store the dataset
train_df.to_csv("train_final.csv", index=False)
test_df.to_csv("test_final.csv", index=False)

print(f"Final number of training set samples: {len(train_df)}")
print(f"Final number of testing set samples: {len(test_df)}")
```

## 3.2 Feature: log returns, volatility, VIX prices & changes

```python
# 2. Feature: log returns & volatility
returns = np.log(prices / prices.shift(1)).dropna()
volatility = returns.rolling(window=20).std().dropna()  # 20 days volatility
returns.rename(columns={'QQQ':'returns'},inplace=True)
volatility.rename(columns={'QQQ':'volatility'},inplace=True)
```

```python
# 2. Feature: Market environment
vix_change = np.log(vix_prices / vix_prices.shift(1))
vix_change.rename(columns={'^VIX':'vix_change'},inplace=True)
vix_prices.rename(columns={'^VIX':"vix"},inplace=True)
```

```python
print(vix_change.columns) ; vix_prices.columns
```

```python
# Concat/Merge
features = pd.concat([returns, volatility,vix_prices,vix_change], axis=1).dropna()
features.head()
```

```python
from hmmlearn import hmm
import numpy as np

# 3. initialize HMM multiple times to get best
def get_stable_hmm(features, n_components=2, n_trials=20):
    best_model = None
    best_score = -np.inf
    for seed in range(n_trials):
        model = hmm.GaussianHMM(n_components=n_components, 
                                covariance_type="full",
                                n_iter=1000, 
                                random_state=seed,
                                verbose=False)

        # 4. Train
        model.fit(features.values)
        score = model.score(features.values)
        if score > best_score:
            best_score = score
            best_model = model
    return best_model

best_hmm = get_stable_hmm(features, n_components=2)

# 5. Predict the most likely state sequence
states = best_hmm.predict(features.values)
```

```python
# 7. Result analysis and visualization
features['state'] = states

# Examine the statistical characteristics of each state to help interpret its meaning later.
state_means = features.groupby('state').agg(['mean', 'std'])
print("The characteristic mean of each state:")
print(state_means)
```

# 5a. Baseline Model

```python
import numpy as np
import pandas as pd
import xgboost as xgb
from sklearn.model_selection import TimeSeriesSplit
from sklearn.metrics import mean_squared_error
from scipy.stats import spearmanr
import warnings
warnings.filterwarnings('ignore')

def evaluate_regression(y_true, y_pred):
    """计算回归指标：MSE 和 IC（Spearman 相关系数）"""
    mse = mean_squared_error(y_true, y_pred)
    ic = spearmanr(y_true, y_pred)[0]
    if np.isnan(ic):
        ic = 0.0
    return mse, ic

def get_default_model_params():
    return {
        'n_estimators': 100,
        'max_depth': 5,
        'learning_rate': 0.05,
        'subsample': 0.8,
        'colsample_bytree': 0.8,
        'random_state': 42,
        'objective': 'reg:squarederror'
    }

def time_series_cv_eval_by_date(X, y, n_splits=5, test_size=60, model_params=None):
    """
    时间序列交叉验证评估 XGBoost 回归模型。
    注意：panel data 必须按日期切分，而不能按行切分，否则同一天的股票会同时落入训练和验证集。
    """
    if model_params is None:
        model_params = get_default_model_params()

    date_index = pd.Index(pd.to_datetime(X.index), name='Date')
    unique_dates = pd.Index(sorted(date_index.unique()))

    if len(unique_dates) <= n_splits * test_size:
        raise ValueError(
            f"Not enough unique dates for CV: got {len(unique_dates)} dates, "
            f"need more than {n_splits * test_size}."
        )

    tscv = TimeSeriesSplit(n_splits=n_splits, test_size=test_size)
    mse_list, ic_list = [], []

    for fold, (train_date_idx, val_date_idx) in enumerate(tscv.split(unique_dates), start=1):
        train_dates = unique_dates[train_date_idx]
        val_dates = unique_dates[val_date_idx]

        train_mask = date_index.isin(train_dates)
        val_mask = date_index.isin(val_dates)

        X_train_fold = X.loc[train_mask]
        y_train_fold = y.loc[train_mask]
        X_val_fold = X.loc[val_mask]
        y_val_fold = y.loc[val_mask]

        model = xgb.XGBRegressor(**model_params)
        model.fit(X_train_fold, y_train_fold)
        y_pred = model.predict(X_val_fold)

        mse, ic = evaluate_regression(y_val_fold, y_pred)
        mse_list.append(mse)
        ic_list.append(ic)

        print(
            f"Fold {fold}: "
            f"train {train_dates.min().date()} -> {train_dates.max().date()} ({train_mask.sum()} rows), "
            f"val {val_dates.min().date()} -> {val_dates.max().date()} ({val_mask.sum()} rows), "
            f"MSE = {mse:.6f}, IC = {ic:.4f}"
        )

    print(f"Average MSE: {np.mean(mse_list):.6f}, Average IC: {np.mean(ic_list):.4f}")
    return np.mean(mse_list), np.mean(ic_list)

def train_final_model(X_train, y_train, model_params=None):
    """在全量训练集上训练最终模型"""
    if model_params is None:
        model_params = get_default_model_params()

    model = xgb.XGBRegressor(**model_params)
    model.fit(X_train, y_train)
    return model
```

```python
# 特征列（包括 state）
feature_cols_with_state = feature_cols + ['state']

# 训练集和测试集
X_train_base = train[feature_cols_with_state].sort_index()
y_train_base = train['future_return'].sort_index()
X_test_base = test[feature_cols_with_state].sort_index()
y_test_base = test['future_return'].sort_index()

# 按日期做时间序列交叉验证（在训练集上）
print("===== Baseline 模型：按日期时间序列交叉验证 =====")
base_cv_mse, base_cv_ic = time_series_cv_eval_by_date(
    X_train_base,
    y_train_base,
    n_splits=5,
    test_size=60
)

# 在全训练集上训练最终模型，并在测试集上评估
print("\n===== Baseline 模型：测试集评估 =====")
base_model = train_final_model(X_train_base, y_train_base)
y_pred_base = base_model.predict(X_test_base)
test_mse, test_ic = evaluate_regression(y_test_base, y_pred_base)
print(f"测试集 MSE: {test_mse:.6f}, IC: {test_ic:.4f}")
```

```python
print("\n===== Regime-aware 模型 =====")

# State 0
X_train_s0 = X_train_state0.sort_index()
y_train_s0 = y_train_state0.sort_index()
X_test_s0 = X_test_state0.sort_index()
y_test_s0 = y_test_state0.sort_index()

if len(X_train_s0) > 0 and len(X_test_s0) > 0:
    print(f"\n--- State 0: 训练样本 {len(X_train_s0)}, 测试样本 {len(X_test_s0)} ---")
    unique_dates_s0 = pd.Index(pd.to_datetime(X_train_s0.index)).nunique()
    if unique_dates_s0 > 120:
        print("State 0 按日期交叉验证:")
        cv_mse_s0, cv_ic_s0 = time_series_cv_eval_by_date(
            X_train_s0,
            y_train_s0,
            n_splits=3,
            test_size=40
        )
    else:
        cv_mse_s0, cv_ic_s0 = np.nan, np.nan
        print("State 0 可用日期不足，跳过交叉验证")

    model_s0 = train_final_model(X_train_s0, y_train_s0)
    y_pred_s0 = model_s0.predict(X_test_s0)
    mse_s0, ic_s0 = evaluate_regression(y_test_s0, y_pred_s0)
    print(f"State 0 测试集 MSE: {mse_s0:.6f}, IC: {ic_s0:.4f}")
else:
    cv_mse_s0, cv_ic_s0 = np.nan, np.nan
    print("State 0 数据不足，跳过")

# State 1
X_train_s1 = X_train_state1.sort_index()
y_train_s1 = y_train_state1.sort_index()
X_test_s1 = X_test_state1.sort_index()
y_test_s1 = y_test_state1.sort_index()

if len(X_train_s1) > 0 and len(X_test_s1) > 0:
    print(f"\n--- State 1: 训练样本 {len(X_train_s1)}, 测试样本 {len(X_test_s1)} ---")
    unique_dates_s1 = pd.Index(pd.to_datetime(X_train_s1.index)).nunique()
    if unique_dates_s1 > 120:
        print("State 1 按日期交叉验证:")
        cv_mse_s1, cv_ic_s1 = time_series_cv_eval_by_date(
            X_train_s1,
            y_train_s1,
            n_splits=3,
            test_size=40
        )
    else:
        cv_mse_s1, cv_ic_s1 = np.nan, np.nan
        print("State 1 可用日期不足，跳过交叉验证")

    model_s1 = train_final_model(X_train_s1, y_train_s1)
    y_pred_s1 = model_s1.predict(X_test_s1)
    mse_s1, ic_s1 = evaluate_regression(y_test_s1, y_pred_s1)
    print(f"State 1 测试集 MSE: {mse_s1:.6f}, IC: {ic_s1:.4f}")
else:
    cv_mse_s1, cv_ic_s1 = np.nan, np.nan
    print("State 1 数据不足，跳过")
```

```python
# 汇总结果
results = pd.DataFrame({
    'Model': ['Baseline (state as feature)', 'Regime-aware State 0', 'Regime-aware State 1'],
    'Test MSE': [test_mse, mse_s0 if len(X_test_s0)>0 else np.nan, mse_s1 if len(X_test_s1)>0 else np.nan],
    'Test IC': [test_ic, ic_s0 if len(X_test_s0)>0 else np.nan, ic_s1 if len(X_test_s1)>0 else np.nan]
})
print("\n===== 最终结果对比 =====")
print(results)
```

# 6. Analysis

对于 cross-sectional stock return prediction，逐样本的 `actual vs predicted` 折线通常不会像单一资产时间序列那样贴合。
更有信息量的诊断是：散点图、按日期聚合后的均值走势、predicted rank 分组收益，以及按日期计算的 cross-sectional IC。

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

def plot_prediction_diagnostics(y_true, y_pred, date_index, title, max_dates=60, n_bins=10):
    diag = pd.DataFrame({
        'Date': pd.to_datetime(date_index),
        'actual': np.asarray(y_true),
        'predicted': np.asarray(y_pred)
    }).dropna()

    fig, axes = plt.subplots(1, 3, figsize=(18, 5))

    # 1. Raw prediction scatter
    axes[0].scatter(diag['actual'], diag['predicted'], alpha=0.25, s=10)
    limits = [
        min(diag['actual'].min(), diag['predicted'].min()),
        max(diag['actual'].max(), diag['predicted'].max())
    ]
    axes[0].plot(limits, limits, 'r--', lw=2, label='Ideal')
    axes[0].set_xlabel('Actual future return')
    axes[0].set_ylabel('Predicted future return')
    axes[0].set_title(f'{title}: scatter')
    axes[0].legend()
    axes[0].grid(True)

    # 2. Date-level average prediction vs actual
    daily_mean = diag.groupby('Date')[['actual', 'predicted']].mean().tail(max_dates)
    axes[1].plot(daily_mean.index, daily_mean['actual'], label='Daily mean actual', alpha=0.8)
    axes[1].plot(daily_mean.index, daily_mean['predicted'], label='Daily mean predicted', alpha=0.8)
    axes[1].set_title(f'{title}: daily mean (last {len(daily_mean)} dates)')
    axes[1].set_xlabel('Date')
    axes[1].set_ylabel('Return')
    axes[1].tick_params(axis='x', rotation=45)
    axes[1].legend()
    axes[1].grid(True)

    # 3. Actual return by predicted rank bin
    if diag['predicted'].nunique() > 1:
        n_bins = min(n_bins, diag['predicted'].nunique())
        diag['pred_bin'] = pd.qcut(diag['predicted'].rank(method='first'), q=n_bins, labels=False)
        rank_bin_return = diag.groupby('pred_bin')['actual'].mean()
        axes[2].bar(np.arange(1, len(rank_bin_return) + 1), rank_bin_return.values)
        axes[2].set_title(f'{title}: actual return by predicted rank')
        axes[2].set_xlabel('Predicted rank bin (low -> high)')
        axes[2].set_ylabel('Mean actual return')
        axes[2].grid(True, axis='y')
    else:
        axes[2].text(0.5, 0.5, 'Not enough prediction variation', ha='center', va='center')
        axes[2].set_axis_off()

    plt.tight_layout()
    plt.show()

print('Note: 对于 panel 回归，逐样本 actual vs predicted 不会像单一时间序列那样紧贴；更应关注 IC、分组收益和按日期聚合后的走势。')

plot_prediction_diagnostics(y_test_s0, y_pred_s0, X_test_s0.index, f'State 0 diagnostics (IC={ic_s0:.3f})')

if 'y_pred_s1' in globals():
    plot_prediction_diagnostics(y_test_s1, y_pred_s1, X_test_s1.index, f'State 1 diagnostics (IC={ic_s1:.3f})')
```

```python
def compute_daily_ic(y_true, y_pred, date_index):
    diag = pd.DataFrame({
        'Date': pd.to_datetime(date_index),
        'actual': np.asarray(y_true),
        'predicted': np.asarray(y_pred)
    }).dropna()

    daily_ic = diag.groupby('Date').apply(
        lambda g: spearmanr(g['actual'], g['predicted'])[0] if g['predicted'].nunique() > 1 else np.nan,
        include_groups=False
    )
    return daily_ic.dropna()

daily_ic_s0 = compute_daily_ic(y_test_s0, y_pred_s0, X_test_s0.index)
print(f'State 0 mean daily IC: {daily_ic_s0.mean():.4f}')

plt.figure(figsize=(12, 4))
plt.plot(daily_ic_s0.index, daily_ic_s0.rolling(20).mean(), label='State 0 rolling 20-day IC')
plt.axhline(0, color='black', linestyle='--', linewidth=1)
plt.title('State 0 daily cross-sectional IC (20-day rolling mean)')
plt.xlabel('Date')
plt.ylabel('IC')
plt.legend()
plt.grid(True)
plt.show()

if 'y_pred_s1' in globals():
    daily_ic_s1 = compute_daily_ic(y_test_s1, y_pred_s1, X_test_s1.index)
    print(f'State 1 mean daily IC: {daily_ic_s1.mean():.4f}')

    plt.figure(figsize=(12, 4))
    plt.plot(daily_ic_s1.index, daily_ic_s1.rolling(20).mean(), label='State 1 rolling 20-day IC', color='tab:orange')
    plt.axhline(0, color='black', linestyle='--', linewidth=1)
    plt.title('State 1 daily cross-sectional IC (20-day rolling mean)')
    plt.xlabel('Date')
    plt.ylabel('IC')
    plt.legend()
    plt.grid(True)
    plt.show()
```

# 7. Monthly Panel, LightGBM, and Fama-MacBeth

这一部分把日频 panel 收敛到 month-end panel，并用更稳的横截面分位数化特征来做月频预测和 Fama-MacBeth。
这样更接近基本面数据的更新频率，也更容易解释 regime 下因子作用是否变化。

```python
import importlib
import subprocess
import sys

try:
    import lightgbm as lgb
except ModuleNotFoundError:
    subprocess.check_call([sys.executable, '-m', 'pip', 'install', 'lightgbm'])
    import lightgbm as lgb

import matplotlib.pyplot as plt
import statsmodels.api as sm
from IPython.display import display
from scipy.stats import t as student_t
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.model_selection import TimeSeriesSplit
from statsmodels.stats.outliers_influence import variance_inflation_factor

def compute_r2_oos(y_true, y_pred, benchmark_mean):
    benchmark_pred = np.full(len(y_true), benchmark_mean, dtype=float)
    numerator = np.sum((np.asarray(y_true) - np.asarray(y_pred)) ** 2)
    denominator = np.sum((np.asarray(y_true) - benchmark_pred) ** 2)
    if denominator == 0:
        return np.nan
    return 1 - numerator / denominator

def evaluate_regression_metrics_monthly(y_true, y_pred, benchmark_mean=None):
    mse = mean_squared_error(y_true, y_pred)
    rmse = np.sqrt(mse)
    mae = mean_absolute_error(y_true, y_pred)
    r2 = r2_score(y_true, y_pred)
    r2_oos = compute_r2_oos(y_true, y_pred, benchmark_mean) if benchmark_mean is not None else np.nan
    ic = spearmanr(y_true, y_pred)[0]
    if np.isnan(ic):
        ic = 0.0
    return {
        'mse': mse,
        'rmse': rmse,
        'mae': mae,
        'r2': r2,
        'r2_oos': r2_oos,
        'ic': ic
    }

def build_month_end_panel(panel, regime_series, feature_cols):
    panel = panel.copy()
    if 'Date' not in panel.columns:
        panel = panel.reset_index()

    panel['Date'] = pd.to_datetime(panel['Date'])
    panel = panel.sort_values(['ticker', 'Date'])

    regime_lookup = regime_series.rename('state').rename_axis('Date').reset_index()
    regime_lookup['Date'] = pd.to_datetime(regime_lookup['Date'])
    panel = panel.merge(regime_lookup, on='Date', how='inner')

    panel['month'] = panel['Date'].dt.to_period('M')
    month_end = panel.groupby(['ticker', 'month'], as_index=False).tail(1).copy()
    month_end = month_end.sort_values(['ticker', 'Date'])
    month_end['future_return_1m'] = month_end.groupby('ticker')['close'].shift(-1) / month_end['close'] - 1
    month_end = month_end.dropna(subset=feature_cols + ['future_return_1m']).reset_index(drop=True)
    return month_end

def cross_sectional_quantile_transform(panel, feature_cols, suffix='_cs_q'):
    panel = panel.copy()

    transformed_groups = []
    for _, df in panel.groupby('Date', sort=False):
        df = df.copy()
        for col in feature_cols:
            ranks = df[col].rank(method='average', pct=True)
            if ranks.nunique() <= 1:
                df[f'{col}{suffix}'] = 0.0
            else:
                df[f'{col}{suffix}'] = (ranks - 0.5) * 2
        transformed_groups.append(df)

    transformed = pd.concat(transformed_groups, ignore_index=True)
    transformed_cols = [f'{col}{suffix}' for col in feature_cols]
    return transformed, transformed_cols

def get_lightgbm_params(objective='regression'):
    return {
        'objective': objective,
        'n_estimators': 300,
        'learning_rate': 0.03,
        'num_leaves': 31,
        'min_child_samples': 10,
        'subsample': 0.8,
        'colsample_bytree': 0.8,
        'random_state': 42,
        'verbosity': -1
    }

def build_lightgbm_model(objective='regression'):
    return lgb.LGBMRegressor(**get_lightgbm_params(objective=objective))

def time_series_cv_eval_model_by_date(X, y, build_model, n_splits=3, test_size=6):
    date_index = pd.Index(pd.to_datetime(X.index), name='Date')
    unique_dates = pd.Index(sorted(date_index.unique()))

    if len(unique_dates) <= n_splits * test_size:
        raise ValueError(
            f'Not enough unique dates for CV: got {len(unique_dates)} dates, '
            f'need more than {n_splits * test_size}.'
        )

    tscv = TimeSeriesSplit(n_splits=n_splits, test_size=test_size)
    fold_metrics = []

    for fold, (train_date_idx, val_date_idx) in enumerate(tscv.split(unique_dates), start=1):
        train_dates = unique_dates[train_date_idx]
        val_dates = unique_dates[val_date_idx]

        train_mask = date_index.isin(train_dates)
        val_mask = date_index.isin(val_dates)

        y_train_fold = y.loc[train_mask]
        model = build_model()
        model.fit(X.loc[train_mask], y_train_fold)
        y_pred = model.predict(X.loc[val_mask])
        metrics = evaluate_regression_metrics_monthly(
            y.loc[val_mask],
            y_pred,
            benchmark_mean=y_train_fold.mean()
        )

        fold_metrics.append(metrics)
        print(
            f"Fold {fold}: "
            f"train {train_dates.min().date()} -> {train_dates.max().date()} ({train_mask.sum()} rows), "
            f"val {val_dates.min().date()} -> {val_dates.max().date()} ({val_mask.sum()} rows), "
            f"MSE = {metrics['mse']:.6f}, RMSE = {metrics['rmse']:.4f}, "
            f"MAE = {metrics['mae']:.4f}, R2 = {metrics['r2']:.4f}, "
            f"R2_oos = {metrics['r2_oos']:.4f}, IC = {metrics['ic']:.4f}"
        )

    summary = pd.DataFrame(fold_metrics).mean().to_dict()
    print(
        f"Average MSE: {summary['mse']:.6f}, Average RMSE: {summary['rmse']:.4f}, "
        f"Average MAE: {summary['mae']:.4f}, Average R2: {summary['r2']:.4f}, "
        f"Average R2_oos: {summary['r2_oos']:.4f}, Average IC: {summary['ic']:.4f}"
    )
    return summary

def fit_regime_aware_models(X_train, y_train, build_model, regime_col='state', base_feature_cols=None):
    if base_feature_cols is None:
        base_feature_cols = [col for col in X_train.columns if col != regime_col]

    fallback_model = build_model()
    fallback_model.fit(X_train[base_feature_cols], y_train)

    state_models = {}
    for state_value in sorted(pd.unique(X_train[regime_col])):
        train_mask = X_train[regime_col] == state_value
        if train_mask.sum() == 0:
            continue
        model = build_model()
        model.fit(X_train.loc[train_mask, base_feature_cols], y_train.loc[train_mask])
        state_models[state_value] = model

    return {
        'state_models': state_models,
        'fallback_model': fallback_model,
        'base_feature_cols': list(base_feature_cols),
        'regime_col': regime_col
    }

def predict_regime_aware_from_bundle(model_bundle, X_test):
    regime_col = model_bundle['regime_col']
    base_feature_cols = model_bundle['base_feature_cols']
    fallback_model = model_bundle['fallback_model']
    state_models = model_bundle['state_models']

    predictions = pd.Series(index=X_test.index, dtype=float)
    for state_value, model in state_models.items():
        test_mask = X_test[regime_col] == state_value
        if test_mask.sum() == 0:
            continue
        predictions.loc[test_mask] = model.predict(X_test.loc[test_mask, base_feature_cols])

    missing_mask = predictions.isna()
    if missing_mask.any():
        predictions.loc[missing_mask] = fallback_model.predict(X_test.loc[missing_mask, base_feature_cols])

    return predictions

def fit_regime_aware_predict(X_train, y_train, X_test, build_model, regime_col='state', base_feature_cols=None):
    model_bundle = fit_regime_aware_models(
        X_train,
        y_train,
        build_model=build_model,
        regime_col=regime_col,
        base_feature_cols=base_feature_cols
    )
    return predict_regime_aware_from_bundle(model_bundle, X_test)

def time_series_cv_eval_regime_aware_by_date(X, y, build_model, regime_col='state', base_feature_cols=None, n_splits=3, test_size=6):
    date_index = pd.Index(pd.to_datetime(X.index), name='Date')
    unique_dates = pd.Index(sorted(date_index.unique()))

    if len(unique_dates) <= n_splits * test_size:
        raise ValueError(
            f'Not enough unique dates for CV: got {len(unique_dates)} dates, '
            f'need more than {n_splits * test_size}.'
        )

    tscv = TimeSeriesSplit(n_splits=n_splits, test_size=test_size)
    fold_metrics = []

    for fold, (train_date_idx, val_date_idx) in enumerate(tscv.split(unique_dates), start=1):
        train_dates = unique_dates[train_date_idx]
        val_dates = unique_dates[val_date_idx]

        train_mask = date_index.isin(train_dates)
        val_mask = date_index.isin(val_dates)

        y_train_fold = y.loc[train_mask]
        fold_pred = fit_regime_aware_predict(
            X.loc[train_mask],
            y_train_fold,
            X.loc[val_mask],
            build_model=build_model,
            regime_col=regime_col,
            base_feature_cols=base_feature_cols
        )
        metrics = evaluate_regression_metrics_monthly(
            y.loc[val_mask],
            fold_pred,
            benchmark_mean=y_train_fold.mean()
        )

        fold_metrics.append(metrics)
        print(
            f"Fold {fold}: "
            f"train {train_dates.min().date()} -> {train_dates.max().date()} ({train_mask.sum()} rows), "
            f"val {val_dates.min().date()} -> {val_dates.max().date()} ({val_mask.sum()} rows), "
            f"MSE = {metrics['mse']:.6f}, RMSE = {metrics['rmse']:.4f}, "
            f"MAE = {metrics['mae']:.4f}, R2 = {metrics['r2']:.4f}, "
            f"R2_oos = {metrics['r2_oos']:.4f}, IC = {metrics['ic']:.4f}"
        )

    summary = pd.DataFrame(fold_metrics).mean().to_dict()
    print(
        f"Average MSE: {summary['mse']:.6f}, Average RMSE: {summary['rmse']:.4f}, "
        f"Average MAE: {summary['mae']:.4f}, Average R2: {summary['r2']:.4f}, "
        f"Average R2_oos: {summary['r2_oos']:.4f}, Average IC: {summary['ic']:.4f}"
    )
    return summary

def compare_lightgbm_objectives(X_train, y_train, X_test, y_test, objectives=None):
    if objectives is None:
        objectives = {
            'L2': 'regression',
            'L1': 'regression_l1',
            'Huber': 'huber'
        }

    rows = []
    benchmark_mean = y_train.mean()
    for label, objective in objectives.items():
        print(f'===== Objective: {label} ({objective}) =====')
        cv_summary = time_series_cv_eval_model_by_date(
            X_train,
            y_train,
            build_model=lambda obj=objective: build_lightgbm_model(objective=obj),
            n_splits=3,
            test_size=6
        )
        model = build_lightgbm_model(objective=objective)
        model.fit(X_train, y_train)
        test_pred = model.predict(X_test)
        test_summary = evaluate_regression_metrics_monthly(y_test, test_pred, benchmark_mean=benchmark_mean)
        rows.append({
            'Objective Label': label,
            'Objective': objective,
            'CV MSE': cv_summary['mse'],
            'CV RMSE': cv_summary['rmse'],
            'CV MAE': cv_summary['mae'],
            'CV R2': cv_summary['r2'],
            'CV R2_oos': cv_summary['r2_oos'],
            'CV IC': cv_summary['ic'],
            'Test MSE': test_summary['mse'],
            'Test RMSE': test_summary['rmse'],
            'Test MAE': test_summary['mae'],
            'Test R2': test_summary['r2'],
            'Test R2_oos': test_summary['r2_oos'],
            'Test IC': test_summary['ic']
        })
    return pd.DataFrame(rows)

def get_feature_group_definitions(feature_cols):
    technical_features = [
        'return_1m', 'return_3m', 'momentum_12m', 'sma_50_sma_200_ratio',
        'MACDh', 'volatility_1m', 'ATR', 'log_volume', 'obv'
    ]
    fundamental_features = [
        'pe_ratio', 'pb_ratio', 'ps_ratio', 'pfcf_ratio',
        'return_on_equity', 'return_on_assets', 'ebitda_margin',
        'debt_to_equity', 'current_ratio'
    ]

    missing = sorted(set(feature_cols) - set(technical_features) - set(fundamental_features))
    if missing:
        raise ValueError(f'Unassigned features found: {missing}')

    return {
        'Technical-only': [col for col in technical_features if col in feature_cols],
        'Fundamental-only': [col for col in fundamental_features if col in feature_cols]
    }

def get_candidate_feature_sets():
    return {
        'technical': [
            'return_5d', 'return_1m', 'return_3m', 'momentum_12m', 'price_vs_sma20',
            'sma_50_sma_200_ratio', 'MACDh', 'adx', 'ATR', 'stdev_feature', 'obv', 'vpt'
        ],
        'fundamental': [
            'pe_ratio', 'pb_ratio', 'ps_ratio', 'pfcf_ratio', 'return_on_equity',
            'return_on_assets', 'ebitda_margin', 'operating_margin', 'gross_margin',
            'quick_ratio', 'current_ratio', 'debt_to_equity', 'cash_to_liabilities'
        ]
    }

def _safe_series(panel, column_name):
    if column_name in panel.columns:
        return panel[column_name]
    return pd.Series(np.nan, index=panel.index, dtype=float)

def add_candidate_features(panel):
    panel = panel.sort_values(['ticker', 'Date']).copy()
    close_group = panel.groupby('ticker')['close']

    panel['return_5d'] = close_group.pct_change(periods=5)
    sma20 = close_group.transform(lambda s: s.rolling(window=20, min_periods=20).mean())
    panel['price_vs_sma20'] = panel['close'] / sma20 - 1
    panel['adx'] = _safe_series(panel, 'adx')
    panel['stdev_feature'] = _safe_series(panel, 'stdev')
    panel['vpt'] = _safe_series(panel, 'vpt')
    panel['quick_ratio'] = _safe_series(panel, 'quick_ratio')

    revenue = _safe_series(panel, 'revenue')
    operating_income = _safe_series(panel, 'operating_income')
    gross_profit = _safe_series(panel, 'gross_profit')
    current_liabilities = _safe_series(panel, 'current_liabilities')
    cash_on_hand = _safe_series(panel, 'cash_on_hand')

    panel['operating_margin'] = np.where(revenue != 0, operating_income / revenue, np.nan)
    panel['gross_margin'] = np.where(revenue != 0, gross_profit / revenue, np.nan)
    panel['cash_to_liabilities'] = np.where(current_liabilities != 0, cash_on_hand / current_liabilities, np.nan)
    return panel

def map_raw_to_cs_features(raw_feature_cols, suffix='_cs_q'):
    return [f'{col}{suffix}' for col in raw_feature_cols]

def extract_feature_importance(model, feature_names, model_label, feature_set_label='All features', normalize=True):
    importance = pd.DataFrame({
        'feature': feature_names,
        'importance': model.feature_importances_
    })
    if normalize and importance['importance'].sum() > 0:
        importance['importance_pct'] = importance['importance'] / importance['importance'].sum()
    else:
        importance['importance_pct'] = importance['importance']
    importance['Model'] = model_label
    importance['Feature Set'] = feature_set_label
    return importance.sort_values('importance', ascending=False).reset_index(drop=True)

def summarize_regime_aware_importance(model_bundle, X_train, feature_set_label='All features'):
    feature_frames = []
    regime_col = model_bundle['regime_col']
    base_feature_cols = model_bundle['base_feature_cols']
    state_counts = X_train[regime_col].value_counts(normalize=True).to_dict()

    for state_value, model in model_bundle['state_models'].items():
        state_df = extract_feature_importance(
            model,
            base_feature_cols,
            model_label=f'Regime-aware state {state_value}',
            feature_set_label=feature_set_label,
            normalize=True
        )
        state_df['state'] = state_value
        state_df['state_weight'] = state_counts.get(state_value, 0.0)
        feature_frames.append(state_df)

    combined = pd.concat(feature_frames, ignore_index=True)
    aggregate = combined.groupby('feature', as_index=False).apply(
        lambda df: pd.Series({
            'importance': np.average(df['importance'], weights=df['state_weight']),
            'importance_pct': np.average(df['importance_pct'], weights=df['state_weight'])
        }),
        include_groups=False
    ).reset_index(drop=True)
    aggregate['Model'] = 'Regime-aware weighted average'
    aggregate['Feature Set'] = feature_set_label
    return combined, aggregate.sort_values('importance_pct', ascending=False).reset_index(drop=True)

def plot_feature_importance_bar(importance_df, title, value_col='importance_pct', top_n=12):
    plot_df = importance_df.sort_values(value_col, ascending=False).head(top_n).copy()
    plot_df = plot_df.sort_values(value_col, ascending=True)

    plt.figure(figsize=(8, 5))
    plt.barh(plot_df['feature'], plot_df[value_col], color='tab:blue')
    plt.title(title)
    plt.xlabel(value_col)
    plt.ylabel('Feature')
    plt.grid(True, axis='x', alpha=0.3)
    plt.tight_layout()
    plt.show()

def compute_univariate_monthly_ic(panel, feature_cols, target_col='future_return_1m'):
    records = []
    for feature in feature_cols:
        per_date_ic = []
        for _, group in panel.groupby('Date'):
            subset = group[[feature, target_col]].dropna()
            if len(subset) < 10:
                continue
            ic = spearmanr(subset[feature], subset[target_col])[0]
            if not np.isnan(ic):
                per_date_ic.append(ic)
        records.append({
            'feature': feature,
            'mean_ic': np.mean(per_date_ic) if per_date_ic else np.nan,
            'abs_mean_ic': np.abs(np.mean(per_date_ic)) if per_date_ic else np.nan
        })
    return pd.DataFrame(records)

def compute_vif_table(panel, feature_cols):
    vif_df = panel[feature_cols].dropna().copy()
    vif_df = vif_df.loc[:, vif_df.nunique() > 1]
    if vif_df.empty:
        return pd.DataFrame(columns=['feature', 'vif'])

    vif_values = []
    for idx, feature in enumerate(vif_df.columns):
        vif_values.append({
            'feature': feature,
            'vif': variance_inflation_factor(vif_df.values, idx)
        })
    return pd.DataFrame(vif_values).sort_values('vif', ascending=False).reset_index(drop=True)

def build_candidate_diagnostic_table(monthly_panel_cs, candidate_feature_cols):
    corr_matrix = monthly_panel_cs[candidate_feature_cols].corr(method='spearman')
    max_abs_corr = []
    for feature in candidate_feature_cols:
        others = corr_matrix[feature].drop(index=feature).abs()
        max_abs_corr.append({
            'feature': feature,
            'max_abs_corr': others.max() if len(others) else np.nan,
            'most_correlated_with': others.idxmax() if len(others) else None
        })
    max_abs_corr = pd.DataFrame(max_abs_corr)

    vif_table = compute_vif_table(monthly_panel_cs, candidate_feature_cols)
    ic_table = compute_univariate_monthly_ic(monthly_panel_cs, candidate_feature_cols)

    diagnostic = max_abs_corr.merge(vif_table, on='feature', how='left').merge(ic_table, on='feature', how='left')
    diagnostic = diagnostic.sort_values(['vif', 'max_abs_corr', 'abs_mean_ic'], ascending=[False, False, False]).reset_index(drop=True)
    return diagnostic, corr_matrix

def run_monthly_triplet_comparison(train_panel, test_panel, raw_feature_cols, label, build_model=build_lightgbm_model):
    cs_feature_cols = map_raw_to_cs_features(raw_feature_cols)
    missing_cols = sorted((set(cs_feature_cols) - set(train_panel.columns)) | (set(cs_feature_cols) - set(test_panel.columns)))
    if missing_cols:
        raise ValueError(f'Missing transformed feature columns: {missing_cols}')

    X_train_pure = train_panel.set_index('Date')[cs_feature_cols].sort_index()
    y_train = train_panel.set_index('Date')['future_return_1m'].sort_index()
    X_test_pure = test_panel.set_index('Date')[cs_feature_cols].sort_index()
    y_test = test_panel.set_index('Date')['future_return_1m'].sort_index()

    X_train_with_state = train_panel.set_index('Date')[cs_feature_cols + ['state']].sort_index()
    X_test_with_state = test_panel.set_index('Date')[cs_feature_cols + ['state']].sort_index()
    benchmark_mean = y_train.mean()

    print(f'===== {label}: pure baseline =====')
    cv_pure = time_series_cv_eval_model_by_date(X_train_pure, y_train, build_model, n_splits=3, test_size=6)
    pure_model = build_model()
    pure_model.fit(X_train_pure, y_train)
    pure_pred = pure_model.predict(X_test_pure)
    test_pure = evaluate_regression_metrics_monthly(y_test, pure_pred, benchmark_mean=benchmark_mean)

    print(f'\n===== {label}: state as feature =====')
    cv_with_state = time_series_cv_eval_model_by_date(X_train_with_state, y_train, build_model, n_splits=3, test_size=6)
    state_feature_model = build_model()
    state_feature_model.fit(X_train_with_state, y_train)
    state_feature_pred = state_feature_model.predict(X_test_with_state)
    test_with_state = evaluate_regression_metrics_monthly(y_test, state_feature_pred, benchmark_mean=benchmark_mean)

    print(f'\n===== {label}: regime-aware =====')
    cv_regime_aware = time_series_cv_eval_regime_aware_by_date(
        X_train_with_state,
        y_train,
        build_model=build_model,
        regime_col='state',
        base_feature_cols=cs_feature_cols,
        n_splits=3,
        test_size=6
    )
    regime_bundle = fit_regime_aware_models(
        X_train_with_state,
        y_train,
        build_model=build_model,
        regime_col='state',
        base_feature_cols=cs_feature_cols
    )
    regime_pred = predict_regime_aware_from_bundle(regime_bundle, X_test_with_state)
    test_regime_aware = evaluate_regression_metrics_monthly(y_test, regime_pred, benchmark_mean=benchmark_mean)

    results = pd.DataFrame([
        {
            'Feature Set': label,
            'Model': 'Monthly LightGBM (pure baseline)',
            'CV MSE': cv_pure['mse'],
            'CV RMSE': cv_pure['rmse'],
            'CV MAE': cv_pure['mae'],
            'CV R2': cv_pure['r2'],
            'CV R2_oos': cv_pure['r2_oos'],
            'CV IC': cv_pure['ic'],
            'Test MSE': test_pure['mse'],
            'Test RMSE': test_pure['rmse'],
            'Test MAE': test_pure['mae'],
            'Test R2': test_pure['r2'],
            'Test R2_oos': test_pure['r2_oos'],
            'Test IC': test_pure['ic']
        },
        {
            'Feature Set': label,
            'Model': 'Monthly LightGBM (state as feature)',
            'CV MSE': cv_with_state['mse'],
            'CV RMSE': cv_with_state['rmse'],
            'CV MAE': cv_with_state['mae'],
            'CV R2': cv_with_state['r2'],
            'CV R2_oos': cv_with_state['r2_oos'],
            'CV IC': cv_with_state['ic'],
            'Test MSE': test_with_state['mse'],
            'Test RMSE': test_with_state['rmse'],
            'Test MAE': test_with_state['mae'],
            'Test R2': test_with_state['r2'],
            'Test R2_oos': test_with_state['r2_oos'],
            'Test IC': test_with_state['ic']
        },
        {
            'Feature Set': label,
            'Model': 'Monthly LightGBM (regime-aware)',
            'CV MSE': cv_regime_aware['mse'],
            'CV RMSE': cv_regime_aware['rmse'],
            'CV MAE': cv_regime_aware['mae'],
            'CV R2': cv_regime_aware['r2'],
            'CV R2_oos': cv_regime_aware['r2_oos'],
            'CV IC': cv_regime_aware['ic'],
            'Test MSE': test_regime_aware['mse'],
            'Test RMSE': test_regime_aware['rmse'],
            'Test MAE': test_regime_aware['mae'],
            'Test R2': test_regime_aware['r2'],
            'Test R2_oos': test_regime_aware['r2_oos'],
            'Test IC': test_regime_aware['ic']
        }
    ])

    model_store = {
        'pure_model': pure_model,
        'state_feature_model': state_feature_model,
        'regime_bundle': regime_bundle,
        'raw_feature_cols': list(raw_feature_cols),
        'cs_feature_cols': list(cs_feature_cols),
        'X_train_pure': X_train_pure,
        'X_train_with_state': X_train_with_state,
        'X_test_pure': X_test_pure,
        'X_test_with_state': X_test_with_state,
        'y_train': y_train,
        'y_test': y_test,
        'benchmark_mean': benchmark_mean,
        'predictions': {
            'pure': pure_pred,
            'state_as_feature': state_feature_pred,
            'regime_aware': regime_pred
        }
    }
    return results, model_store

def run_fama_macbeth(panel, feature_cols, target_col='future_return_1m', min_obs=25):
    beta_frames = []

    for date, group in panel.groupby('Date'):
        reg_data = group[feature_cols + [target_col]].dropna()
        if len(reg_data) < max(min_obs, len(feature_cols) + 2):
            continue

        X = sm.add_constant(reg_data[feature_cols], has_constant='add')
        y = reg_data[target_col]
        result = sm.OLS(y, X).fit()

        params = result.params
        params.name = date
        beta_frames.append(params)

    if not beta_frames:
        raise ValueError('No monthly cross-sections met the minimum observation requirement.')

    betas = pd.DataFrame(beta_frames)
    mean_beta = betas.mean()
    se_beta = betas.std(ddof=1) / np.sqrt(len(betas))
    t_stat = mean_beta / se_beta.replace(0, np.nan)
    p_value = 2 * (1 - student_t.cdf(np.abs(t_stat), df=max(len(betas) - 1, 1)))

    summary = pd.DataFrame({
        'mean_beta': mean_beta,
        't_stat': t_stat,
        'p_value': p_value
    })
    summary['abs_t'] = summary['t_stat'].abs()
    summary = summary.sort_values('abs_t', ascending=False)
    return summary, betas
```

# 8. Monthly Walk-Forward Validator

这一部分把旧的单序列 validator 改成适用于 month-end panel 的版本。
核心区别是按月份切分、每个月做横截面排序评估，并把 IC 与 long-short spread 作为主指标。

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from scipy.stats import spearmanr

class MonthlyPanelWalkForwardValidator:
    def __init__(self, panel, date_col='Date', ticker_col='ticker', target_col='future_return_1m'):
        self.date_col = date_col
        self.ticker_col = ticker_col
        self.target_col = target_col

        self.panel = panel.copy()
        self.panel[self.date_col] = pd.to_datetime(self.panel[self.date_col])
        self.panel = self.panel.sort_values([self.date_col, self.ticker_col]).reset_index(drop=True)
        self.unique_dates = pd.Index(sorted(self.panel[self.date_col].unique()))

    def _compute_cross_sectional_metrics(self, scored_df, n_bins=5):
        per_date_records = []

        for date, group in scored_df.groupby(self.date_col):
            if len(group) < max(n_bins, 5):
                continue

            ic = spearmanr(group[self.target_col], group['prediction'])[0]
            if np.isnan(ic):
                ic = 0.0

            ranked = group.copy()
            ranked['pred_bin'] = pd.qcut(
                ranked['prediction'].rank(method='first'),
                q=min(n_bins, len(ranked)),
                labels=False,
                duplicates='drop'
            )

            top_mean = ranked.loc[ranked['pred_bin'] == ranked['pred_bin'].max(), self.target_col].mean()
            bottom_mean = ranked.loc[ranked['pred_bin'] == ranked['pred_bin'].min(), self.target_col].mean()

            per_date_records.append({
                self.date_col: date,
                'ic': ic,
                'top_mean_return': top_mean,
                'bottom_mean_return': bottom_mean,
                'long_short_spread': top_mean - bottom_mean,
                'n_stocks': len(group)
            })

        if not per_date_records:
            return pd.DataFrame(columns=[self.date_col, 'ic', 'top_mean_return', 'bottom_mean_return', 'long_short_spread', 'n_stocks'])

        return pd.DataFrame(per_date_records)

    def walk_forward_validation(
        self,
        build_model,
        feature_cols,
        initial_train_months=48,
        step_size=3,
        mode='expanding',
        window_size=36,
        n_bins=5
    ):
        fold_predictions = []
        fold_summaries = []

        if mode not in {'expanding', 'rolling'}:
            raise ValueError("mode must be 'expanding' or 'rolling'")

        train_end = initial_train_months
        fold = 0

        while train_end < len(self.unique_dates):
            if mode == 'expanding':
                train_dates = self.unique_dates[:train_end]
            else:
                train_start = max(0, train_end - window_size)
                train_dates = self.unique_dates[train_start:train_end]

            test_dates = self.unique_dates[train_end:min(train_end + step_size, len(self.unique_dates))]
            if len(test_dates) == 0:
                break

            train_df = self.panel[self.panel[self.date_col].isin(train_dates)].copy()
            test_df = self.panel[self.panel[self.date_col].isin(test_dates)].copy()

            X_train = train_df[feature_cols]
            y_train = train_df[self.target_col]
            X_test = test_df[feature_cols]
            y_test = test_df[self.target_col]

            model = build_model()
            model.fit(X_train, y_train)
            y_pred = model.predict(X_test)
            benchmark_mean = y_train.mean()

            keep_cols = [self.date_col, self.ticker_col, self.target_col]
            if 'state' in test_df.columns:
                keep_cols.append('state')
            scored_df = test_df[keep_cols].copy()
            scored_df['prediction'] = y_pred
            scored_df['fold'] = fold
            scored_df['mode'] = mode

            cs_metrics = self._compute_cross_sectional_metrics(scored_df, n_bins=n_bins)

            fold_summary = {
                'fold': fold,
                'mode': mode,
                'train_months': len(train_dates),
                'test_months': len(test_dates),
                'train_rows': len(train_df),
                'test_rows': len(test_df),
                'train_start': pd.Timestamp(train_dates.min()),
                'train_end': pd.Timestamp(train_dates.max()),
                'test_start': pd.Timestamp(test_dates.min()),
                'test_end': pd.Timestamp(test_dates.max()),
                'test_mae': mean_absolute_error(y_test, y_pred),
                'test_mse': mean_squared_error(y_test, y_pred),
                'test_rmse': np.sqrt(mean_squared_error(y_test, y_pred)),
                'test_r2': r2_score(y_test, y_pred),
                'test_r2_oos': compute_r2_oos(y_test, y_pred, benchmark_mean),
                'mean_ic': cs_metrics['ic'].mean() if len(cs_metrics) else np.nan,
                'mean_long_short_spread': cs_metrics['long_short_spread'].mean() if len(cs_metrics) else np.nan
            }

            print(
                f"Fold {fold} ({mode}): train {fold_summary['train_start'].date()} -> {fold_summary['train_end'].date()} "
                f"({fold_summary['train_rows']} rows), test {fold_summary['test_start'].date()} -> {fold_summary['test_end'].date()} "
                f"({fold_summary['test_rows']} rows)"
            )
            print(
                f"    MSE={fold_summary['test_mse']:.6f}, RMSE={fold_summary['test_rmse']:.4f}, "
                f"R2={fold_summary['test_r2']:.4f}, R2_oos={fold_summary['test_r2_oos']:.4f}, "
                f"IC={fold_summary['mean_ic']:.4f}, LS Spread={fold_summary['mean_long_short_spread']:.4f}\n"
            )

            fold_predictions.append(scored_df)
            fold_summaries.append(fold_summary)

            train_end += step_size
            fold += 1

        predictions_df = pd.concat(fold_predictions, ignore_index=True) if fold_predictions else pd.DataFrame()
        fold_summary_df = pd.DataFrame(fold_summaries)
        return predictions_df, fold_summary_df

    def plot_prediction_scatter(self, predictions_df, title):
        plt.figure(figsize=(7, 5))
        plt.scatter(predictions_df[self.target_col], predictions_df['prediction'], alpha=0.3, s=18)
        low = min(predictions_df[self.target_col].min(), predictions_df['prediction'].min())
        high = max(predictions_df[self.target_col].max(), predictions_df['prediction'].max())
        plt.plot([low, high], [low, high], 'r--', linewidth=2)
        plt.xlabel('Actual return')
        plt.ylabel('Predicted return')
        plt.title(title)
        plt.grid(True)
        plt.show()

    def plot_long_short_curve(self, predictions_df, title, n_bins=5):
        cs_metrics = self._compute_cross_sectional_metrics(predictions_df, n_bins=n_bins)
        if cs_metrics.empty:
            print('No cross-sectional metrics available for plotting.')
            return

        cs_metrics = cs_metrics.sort_values(self.date_col).copy()
        cs_metrics['cum_long_short'] = (1 + cs_metrics['long_short_spread']).cumprod() - 1

        fig, axes = plt.subplots(1, 2, figsize=(14, 4))
        axes[0].plot(cs_metrics[self.date_col], cs_metrics['ic'], marker='o')
        axes[0].axhline(0, color='black', linestyle='--', linewidth=1)
        axes[0].set_title(f'{title}: monthly IC')
        axes[0].set_xlabel('Date')
        axes[0].set_ylabel('IC')
        axes[0].tick_params(axis='x', rotation=45)
        axes[0].grid(True)

        axes[1].plot(cs_metrics[self.date_col], cs_metrics['cum_long_short'], marker='o', color='tab:green')
        axes[1].axhline(0, color='black', linestyle='--', linewidth=1)
        axes[1].set_title(f'{title}: cumulative long-short spread')
        axes[1].set_xlabel('Date')
        axes[1].set_ylabel('Cumulative return')
        axes[1].tick_params(axis='x', rotation=45)
        axes[1].grid(True)

        plt.tight_layout()
        plt.show()

    def plot_decile_bar_by_regime(self, predictions_df, title, regime_col='state', n_bins=10):
        if regime_col not in predictions_df.columns:
            print(f'{regime_col} is not available in predictions_df.')
            return

        plot_df = predictions_df.copy()
        grouped_bars = []
        spread_rows = []

        for regime_value, regime_df in plot_df.groupby(regime_col):
            ranked_df = regime_df.copy()
            ranked_df['pred_bin'] = ranked_df.groupby(self.date_col)['prediction'].transform(
                lambda s: pd.qcut(s.rank(method='first'), q=min(n_bins, len(s)), labels=False, duplicates='drop')
            )
            ranked_df = ranked_df.dropna(subset=['pred_bin']).copy()
            ranked_df['pred_bin'] = ranked_df['pred_bin'].astype(int) + 1

            decile_means = ranked_df.groupby('pred_bin')[self.target_col].mean().reset_index()
            decile_means[regime_col] = regime_value
            grouped_bars.append(decile_means)

            top_bottom_by_date = ranked_df.groupby([self.date_col, 'pred_bin'])[self.target_col].mean().unstack()
            top_col = top_bottom_by_date.columns.max()
            bottom_col = top_bottom_by_date.columns.min()
            spread_rows.append({
                regime_col: regime_value,
                'avg_top_minus_bottom': (top_bottom_by_date[top_col] - top_bottom_by_date[bottom_col]).mean()
            })

        if not grouped_bars:
            print('Not enough data to build regime decile chart.')
            return

        decile_plot = pd.concat(grouped_bars, ignore_index=True)
        spread_df = pd.DataFrame(spread_rows)

        fig, axes = plt.subplots(1, 2, figsize=(15, 5))
        for regime_value, regime_df in decile_plot.groupby(regime_col):
            axes[0].plot(regime_df['pred_bin'], regime_df[self.target_col], marker='o', label=f'Regime {regime_value}')
        axes[0].set_title(f'{title}: decile mean return by regime')
        axes[0].set_xlabel('Predicted decile (low -> high)')
        axes[0].set_ylabel('Mean actual return')
        axes[0].grid(True)
        axes[0].legend()

        axes[1].bar(spread_df[regime_col].astype(str), spread_df['avg_top_minus_bottom'])
        axes[1].axhline(0, color='black', linestyle='--', linewidth=1)
        axes[1].set_title(f'{title}: top-minus-bottom by regime')
        axes[1].set_xlabel('Regime')
        axes[1].set_ylabel('Average top-bottom spread')
        axes[1].grid(True, axis='y')

        plt.tight_layout()
        plt.show()
```

```python
monthly_validator = MonthlyPanelWalkForwardValidator(
    monthly_train_cs,
    date_col='Date',
    ticker_col='ticker',
    target_col='future_return_1m'
)

validator_feature_cols = monthly_feature_cols_with_state

print('===== Monthly expanding walk-forward =====')
expanding_predictions, expanding_summary = monthly_validator.walk_forward_validation(
    build_model=build_lightgbm_model,
    feature_cols=validator_feature_cols,
    initial_train_months=48,
    step_size=3,
    mode='expanding',
    window_size=36,
    n_bins=5
)

print('===== Monthly rolling walk-forward =====')
rolling_predictions, rolling_summary = monthly_validator.walk_forward_validation(
    build_model=build_lightgbm_model,
    feature_cols=validator_feature_cols,
    initial_train_months=48,
    step_size=3,
    mode='rolling',
    window_size=36,
    n_bins=5
)

validator_comparison = pd.DataFrame([
    {
        'mode': 'expanding',
        'mean_mae': expanding_summary['test_mae'].mean(),
        'mean_mse': expanding_summary['test_mse'].mean(),
        'mean_rmse': expanding_summary['test_rmse'].mean(),
        'mean_r2': expanding_summary['test_r2'].mean(),
        'mean_r2_oos': expanding_summary['test_r2_oos'].mean(),
        'mean_ic': expanding_summary['mean_ic'].mean(),
        'mean_long_short_spread': expanding_summary['mean_long_short_spread'].mean()
    },
    {
        'mode': 'rolling',
        'mean_mae': rolling_summary['test_mae'].mean(),
        'mean_mse': rolling_summary['test_mse'].mean(),
        'mean_rmse': rolling_summary['test_rmse'].mean(),
        'mean_r2': rolling_summary['test_r2'].mean(),
        'mean_r2_oos': rolling_summary['test_r2_oos'].mean(),
        'mean_ic': rolling_summary['mean_ic'].mean(),
        'mean_long_short_spread': rolling_summary['mean_long_short_spread'].mean()
    }
])

print('===== Walk-forward comparison =====')
print('Note: MAPE is intentionally excluded because monthly stock returns can be close to zero or negative.')
display(validator_comparison)
display(expanding_summary)
display(rolling_summary)

monthly_validator.plot_prediction_scatter(expanding_predictions, 'Monthly expanding: actual vs predicted')
monthly_validator.plot_long_short_curve(expanding_predictions, 'Monthly expanding', n_bins=5)
monthly_validator.plot_prediction_scatter(rolling_predictions, 'Monthly rolling: actual vs predicted')
monthly_validator.plot_long_short_curve(rolling_predictions, 'Monthly rolling', n_bins=5)
```

# 13. Fundamental-only Fama-MacBeth Interaction Test

这一部分用 `fundamental-only` 特征做更正式的 Fama-MacBeth 检验。这里采用 `factor × regime` interaction 形式：把 `state = 1` 视为相对 `state = 0` 的 slope shift，若交互项显著，则表示该基本面因子的定价/预测斜率在不同 regime 下确实不同。

```python
fundamental_raw_cols = feature_group_definitions['Fundamental-only']
fundamental_cs_cols = map_raw_to_cs_features(fundamental_raw_cols)

monthly_full_cs = pd.concat([monthly_train_cs, monthly_test_cs], ignore_index=True)
monthly_full_cs = monthly_full_cs.sort_values(['Date', 'ticker']).reset_index(drop=True)

fundamental_fm_panel = monthly_full_cs[[
    'Date', 'ticker', 'state', 'future_return_1m', *fundamental_cs_cols
]].dropna().copy()
fundamental_fm_panel['state_indicator'] = (fundamental_fm_panel['state'] == 1).astype(int)

fundamental_interaction_cols = []
for col in fundamental_cs_cols:
    interaction_col = f'{col}_x_state1'
    fundamental_fm_panel[interaction_col] = fundamental_fm_panel[col] * fundamental_fm_panel['state_indicator']
    fundamental_interaction_cols.append(interaction_col)

fundamental_fm_feature_cols = fundamental_cs_cols + ['state_indicator'] + fundamental_interaction_cols
fm_fundamental_interaction_summary, fm_fundamental_interaction_betas = run_fama_macbeth(
    fundamental_fm_panel,
    fundamental_fm_feature_cols,
    target_col='future_return_1m',
    min_obs=25
)

fm_fundamental_interactions_only = fm_fundamental_interaction_summary.loc[
    ['state_indicator'] + fundamental_interaction_cols
].copy()
fm_fundamental_interactions_only['term_type'] = ['state_level_shift'] + ['factor_x_regime1'] * len(fundamental_interaction_cols)
fm_fundamental_interactions_only = fm_fundamental_interactions_only.sort_values('abs_t', ascending=False)

print('===== Fundamental-only Fama-MacBeth with regime interactions =====')
print('Interpretation: each *_x_state1 coefficient measures how the slope in regime 1 differs from regime 0.')
display(fm_fundamental_interactions_only)

fm_fundamental_main_effects = fm_fundamental_interaction_summary.loc[fundamental_cs_cols].copy()
print('===== Fundamental-only Fama-MacBeth main effects (regime 0 baseline slopes) =====')
display(fm_fundamental_main_effects.sort_values('abs_t', ascending=False))
```

```python
# 15. Correlation, VIF, and Candidate Feature Diagnostics
candidate_feature_sets = get_candidate_feature_sets()
all_candidate_features = candidate_feature_sets['technical'] + candidate_feature_sets['fundamental']

train_enhanced = train_df.copy()
test_enhanced = test_df.copy()
if 'Date' not in train_enhanced.columns:
    train_enhanced = train_enhanced.reset_index()
if 'Date' not in test_enhanced.columns:
    test_enhanced = test_enhanced.reset_index()

train_enhanced = add_candidate_features(train_enhanced)
test_enhanced = add_candidate_features(test_enhanced)

expanded_feature_pool = list(dict.fromkeys(feature_cols + all_candidate_features))
available_expanded_features = [
    col for col in expanded_feature_pool
    if col in train_enhanced.columns and train_enhanced[col].notna().sum() > 0
]
missing_expanded_features = sorted(set(expanded_feature_pool) - set(available_expanded_features))

feature_family_lookup = {}
for col in feature_cols:
    feature_family_lookup[col] = 'Current set'
for col in candidate_feature_sets['technical']:
    feature_family_lookup[col] = 'Technical candidate'
for col in candidate_feature_sets['fundamental']:
    feature_family_lookup[col] = 'Fundamental candidate'

feature_availability = pd.DataFrame({
    'feature': expanded_feature_pool,
    'family': [feature_family_lookup.get(col, 'Other') for col in expanded_feature_pool],
    'available_in_train': [col in available_expanded_features for col in expanded_feature_pool],
    'non_null_train_obs': [train_enhanced[col].notna().sum() if col in train_enhanced.columns else 0 for col in expanded_feature_pool]
}).sort_values(['available_in_train', 'family', 'feature'], ascending=[False, True, True]).reset_index(drop=True)

display(feature_availability)
if missing_expanded_features:
    print('Unavailable or empty candidate features:', missing_expanded_features)

monthly_diagnostic_train_raw = build_month_end_panel(train_enhanced, regime_df, available_expanded_features)
monthly_diagnostic_test_raw = build_month_end_panel(test_enhanced, regime_df, available_expanded_features)
monthly_diagnostic_train_cs, diagnostic_cs_cols = cross_sectional_quantile_transform(monthly_diagnostic_train_raw, available_expanded_features)
monthly_diagnostic_test_cs, _ = cross_sectional_quantile_transform(monthly_diagnostic_test_raw, available_expanded_features)

expanded_feature_diagnostic, expanded_feature_corr = build_candidate_diagnostic_table(monthly_diagnostic_train_cs, diagnostic_cs_cols)
expanded_feature_diagnostic['feature_raw'] = expanded_feature_diagnostic['feature'].str.replace('_cs_q', '', regex=False)
expanded_feature_diagnostic['most_correlated_with_raw'] = expanded_feature_diagnostic['most_correlated_with'].str.replace('_cs_q', '', regex=False)
expanded_feature_diagnostic['family'] = expanded_feature_diagnostic['feature_raw'].map(feature_family_lookup)
expanded_feature_diagnostic = expanded_feature_diagnostic[[
    'feature_raw', 'family', 'vif', 'max_abs_corr', 'most_correlated_with_raw', 'mean_ic', 'abs_mean_ic'
]].rename(columns={
    'feature_raw': 'feature',
    'most_correlated_with_raw': 'most_correlated_with'
}).sort_values(['vif', 'max_abs_corr', 'abs_mean_ic'], ascending=[False, False, False]).reset_index(drop=True)

display(expanded_feature_diagnostic)

plot_corr = expanded_feature_corr.copy()
plot_corr.index = plot_corr.index.str.replace('_cs_q', '', regex=False)
plot_corr.columns = plot_corr.columns.str.replace('_cs_q', '', regex=False)

plt.figure(figsize=(14, 11))
plt.imshow(plot_corr, cmap='coolwarm', vmin=-1, vmax=1)
plt.colorbar(label='Spearman correlation')
plt.xticks(range(len(plot_corr.columns)), plot_corr.columns, rotation=90, fontsize=8)
plt.yticks(range(len(plot_corr.index)), plot_corr.index, fontsize=8)
plt.title('Expanded Feature Pool Correlation Heatmap')
plt.tight_layout()
plt.show()

high_vif_or_corr = expanded_feature_diagnostic[
    (expanded_feature_diagnostic['vif'] >= 5) |
    (expanded_feature_diagnostic['max_abs_corr'] >= 0.8)
].copy()

signal_candidates = expanded_feature_diagnostic.sort_values('abs_mean_ic', ascending=False).head(12).copy()

print('Potential redundancy flags')
display(high_vif_or_corr)
print('Top features by absolute univariate monthly IC')
display(signal_candidates)
# 16. Cleaner Feature Set Re-test
technical_cleaner_preferred = [
    'return_1m', 'momentum_12m', 'MACDh', 'adx', 'ATR', 'obv'
]
fundamental_cleaner_preferred = [
    'pfcf_ratio', 'return_on_assets', 'ebitda_margin', 'gross_margin',
    'quick_ratio', 'debt_to_equity', 'cash_to_liabilities'
]

cleaner_technical_cols = [col for col in technical_cleaner_preferred if col in available_expanded_features]
cleaner_fundamental_cols = [col for col in fundamental_cleaner_preferred if col in available_expanded_features]
cleaner_all_cols = cleaner_technical_cols + cleaner_fundamental_cols

print('Cleaner technical features:', cleaner_technical_cols)
print('Cleaner fundamental features:', cleaner_fundamental_cols)
print('Cleaner all features:', cleaner_all_cols)

monthly_clean_train_raw = build_month_end_panel(train_enhanced, regime_df, cleaner_all_cols)
monthly_clean_test_raw = build_month_end_panel(test_enhanced, regime_df, cleaner_all_cols)
monthly_clean_train_cs, monthly_clean_cs_cols = cross_sectional_quantile_transform(monthly_clean_train_raw, cleaner_all_cols)
monthly_clean_test_cs, _ = cross_sectional_quantile_transform(monthly_clean_test_raw, cleaner_all_cols)

cleaner_feature_set_results = []
cleaner_feature_set_model_store = {}
for feature_set_label, raw_cols in {
    'Cleaner all features': cleaner_all_cols,
    'Cleaner technical-only': cleaner_technical_cols,
    'Cleaner fundamental-only': cleaner_fundamental_cols
}.items():
    results_df, store = run_monthly_triplet_comparison(
        monthly_clean_train_cs,
        monthly_clean_test_cs,
        raw_cols,
        label=feature_set_label
    )
    cleaner_feature_set_results.append(results_df)
    cleaner_feature_set_model_store[feature_set_label] = store

cleaner_feature_set_comparison = pd.concat(cleaner_feature_set_results, ignore_index=True)
display(cleaner_feature_set_comparison)

cleaner_feature_set_test_ic = cleaner_feature_set_comparison.pivot(index='Feature Set', columns='Model', values='Test IC')
cleaner_feature_set_test_r2_oos = cleaner_feature_set_comparison.pivot(index='Feature Set', columns='Model', values='Test R2_oos')
cleaner_feature_set_test_mse = cleaner_feature_set_comparison.pivot(index='Feature Set', columns='Model', values='Test MSE')

print('Cleaner set: Test IC')
display(cleaner_feature_set_test_ic)
print('Cleaner set: Test R2_oos')
display(cleaner_feature_set_test_r2_oos)
print('Cleaner set: Test MSE')
display(cleaner_feature_set_test_mse)

cleaner_feature_summary = pd.DataFrame({
    'bucket': ['technical', 'fundamental', 'all'],
    'feature_count': [len(cleaner_technical_cols), len(cleaner_fundamental_cols), len(cleaner_all_cols)],
    'train_rows': [len(monthly_clean_train_cs), len(monthly_clean_train_cs), len(monthly_clean_train_cs)],
    'test_rows': [len(monthly_clean_test_cs), len(monthly_clean_test_cs), len(monthly_clean_test_cs)]
})
display(cleaner_feature_summary)
```

# 17. Cleaner Fundamental-only Fama-MacBeth Interaction Test

```python
# 17. Cleaner Fundamental-only Fama-MacBeth Interaction Test
# Same factor-by-regime interaction design as before, but with the cleaner fundamental-only feature set.
cleaner_fundamental_cs_cols = map_raw_to_cs_features(cleaner_fundamental_cols)

monthly_clean_full_cs = pd.concat([monthly_clean_train_cs, monthly_clean_test_cs], ignore_index=True)
monthly_clean_full_cs = monthly_clean_full_cs.sort_values(['Date', 'ticker']).reset_index(drop=True)

cleaner_fundamental_fm_panel = monthly_clean_full_cs[[
    'Date', 'ticker', 'state', 'future_return_1m', *cleaner_fundamental_cs_cols
]].dropna().copy()
cleaner_fundamental_fm_panel['state_indicator'] = (cleaner_fundamental_fm_panel['state'] == 1).astype(int)

cleaner_fundamental_interaction_cols = []
for col in cleaner_fundamental_cs_cols:
    interaction_col = f'{col}_x_state1'
    cleaner_fundamental_fm_panel[interaction_col] = (
        cleaner_fundamental_fm_panel[col] * cleaner_fundamental_fm_panel['state_indicator']
    )
    cleaner_fundamental_interaction_cols.append(interaction_col)

cleaner_fundamental_fm_feature_cols = (
    cleaner_fundamental_cs_cols + ['state_indicator'] + cleaner_fundamental_interaction_cols
)
cleaner_fm_fundamental_interaction_summary, cleaner_fm_fundamental_interaction_betas = run_fama_macbeth(
    cleaner_fundamental_fm_panel,
    cleaner_fundamental_fm_feature_cols,
    target_col='future_return_1m',
    min_obs=25
)

cleaner_fm_fundamental_interactions_only = cleaner_fm_fundamental_interaction_summary.loc[
    ['state_indicator'] + cleaner_fundamental_interaction_cols
].copy()
cleaner_fm_fundamental_interactions_only['term_type'] = [
    'state_level_shift'
] + ['factor_x_regime1'] * len(cleaner_fundamental_interaction_cols)
cleaner_fm_fundamental_interactions_only['feature'] = [
    'state_indicator'
] + cleaner_fundamental_cols
cleaner_fm_fundamental_interactions_only = cleaner_fm_fundamental_interactions_only.sort_values('abs_t', ascending=False)

cleaner_fm_fundamental_main_effects = cleaner_fm_fundamental_interaction_summary.loc[
    cleaner_fundamental_cs_cols
].copy()
cleaner_fm_fundamental_main_effects['feature'] = cleaner_fundamental_cols
cleaner_fm_fundamental_main_effects = cleaner_fm_fundamental_main_effects.sort_values('abs_t', ascending=False)

original_fm_interactions_named = fm_fundamental_interactions_only.copy()
original_fm_interactions_named['feature'] = ['state_indicator'] + fundamental_raw_cols

cleaner_fm_interaction_comparison = pd.concat([
    original_fm_interactions_named.assign(spec='Original fundamental-only'),
    cleaner_fm_fundamental_interactions_only.assign(spec='Cleaner fundamental-only')
], ignore_index=False)
cleaner_fm_interaction_comparison = cleaner_fm_interaction_comparison[[
    'spec', 'feature', 'term_type', 'mean_beta', 't_stat', 'p_value', 'abs_t'
]].sort_values(['spec', 'abs_t'], ascending=[True, False]).reset_index(drop=True)

print('===== Cleaner fundamental-only Fama-MacBeth with regime interactions =====')
print('Interpretation: each *_x_state1 coefficient measures how the slope in regime 1 differs from regime 0.')
display(cleaner_fm_fundamental_interactions_only[[
    'feature', 'term_type', 'mean_beta', 't_stat', 'p_value', 'abs_t'
]])

print('===== Cleaner fundamental-only Fama-MacBeth main effects (regime 0 baseline slopes) =====')
display(cleaner_fm_fundamental_main_effects[[
    'feature', 'mean_beta', 't_stat', 'p_value', 'abs_t'
]])

print('===== Original vs cleaner interaction-term comparison =====')
display(cleaner_fm_interaction_comparison)
```

# 18. Conclusion and Implications

This project studies whether the predictive power of stock-level signals varies across market regimes, with particular attention to the relative roles of technical and fundamental features. The results suggest that regime information matters, but its contribution is not uniform across feature families. Instead, the evidence points to a comparative pattern: technical and fundamental signals play different predictive roles, and their relative usefulness changes with market conditions.

The technical feature set appears to be more effective as a short-horizon ranking and timing signal. This is consistent with its stronger role in cross-sectional ordering rather than in stable level prediction. By contrast, the cleaner fundamental-only specification becomes more informative once regime information is incorporated. Within the cleaner feature set, the fundamental-only model with state included as a feature delivers the strongest test Information Coefficient and the only positive out-of-sample $R^2$ among the cleaner specifications. This suggests that a more disciplined and less redundant set of fundamental variables can provide a more coherent predictive signal once the market environment is taken into account.

The cleaner fundamental-only Fama-MacBeth interaction test reinforces this interpretation. The regime level shift remains strongly significant, indicating that market state has an important effect on expected cross-sectional returns. At the same time, the interaction results become more concentrated after feature cleaning. Rather than producing many weak and noisy interaction terms, the cleaner specification isolates a more interpretable result, with `cash_to_liabilities` showing the clearest regime-dependent slope difference. This is a stronger and more economically meaningful conclusion than the broader fundamental specification, where interaction effects were more diffuse and less stable.

Taken together, the findings suggest that the main contribution of the project is not simply that regime matters, but that regime changes the relative predictive usefulness of technical and fundamental information. Technical signals appear more useful when the objective is short-run ranking, while cleaner fundamental signals become more attractive when the goal is interpretable regime-conditioned prediction. A practical implication is that feature selection should be conditioned on market status. In some market states, investors may place greater weight on technical indicators for timing and ordering, whereas in others, cleaner balance-sheet and valuation-related fundamentals may offer a more robust basis for cross-sectional prediction.

Overall, the results support a comparative view of predictive power across feature families and regimes. Rather than asking whether fundamentals or technicals are universally better, the more relevant question is which type of information is more informative under a given market regime. This framing provides a clearer foundation for feature design, model interpretation, and future extensions of regime-dependent stock return prediction.
